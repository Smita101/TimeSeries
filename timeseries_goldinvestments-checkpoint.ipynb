```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model=ARIMA(df1['Oil Investments'],order=(1,1,1)) 
model_fit=model.fit()
```

```python
model_fit.summary()
```

```python
df['forecast']=model_fit.predict(start=1800,end=1980,dynamic=True)
df[['Oil Investments','forecast']].plot(figsize=(50,50))
```
